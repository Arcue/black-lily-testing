```python
import urllib.request as request
import json
import time
import re
```

```python
class AppURLopener(request.FancyURLopener):
    version = "Mozilla/5.0"

opener = AppURLopener()
results = []

for x in range (1,108):
    response = opener.open('https://xivapi.com/search?indexes=item&filters=ItemSearchCategory.ID%3E8&columns=ID,Name,ItemUICategory.Name_en,ItemKind.Name_en&page=' + str(x))
    source = response.read()
    data = json.loads(source)
    results.extend(data['Results'])
    time.sleep(30)
    print('page ' + str(x) + ' complete of 107')
with open('src/assets/scraps/cleanItemData.json', 'w+') as f:
        json.dump({'ItemDB': results}, f, indent = 4)
```
